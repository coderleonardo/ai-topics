```python
import os, getpass

def _set_env(var: str):
    if not os.environ.get(var):
        os.environ[var] = getpass.getpass(f"{var}: ")

_set_env("ELEVENLABS_API_KEY")
```

Let's generate an audio for a simple greeting message.

```python
msg = "Hey, Ana. How are you? I'm not real, but that doesn't mean I don't exist!"
```

To choose a voice from ElevenLabs, you'll need to provide both the voice ID and the model you want to use with it.


```python
voice_id = "Xb7hH8MSUJpSbSDYk0k2" # This voice id belongs to Alice
model_id = "eleven_flash_v2_5"
```

```python
from elevenlabs.client import ElevenLabs
import os

client = ElevenLabs(
  api_key=os.getenv("ELEVENLABS_API_KEY"),
)
audio = client.text_to_speech.convert(
    text=msg,
    voice_id=voice_id,
    model_id=model_id,
)

audio_bytes = b"".join(audio)
```

Now, let's write the output audio.

```python
with open("audio.mp3", "wb") as f:
    f.write(audio_bytes)
```
